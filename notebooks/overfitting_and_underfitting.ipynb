```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

# Generate synthetic data
np.random.seed(0)
X = np.random.rand(100, 1) * 10  # Random data points
y = np.sin(X).ravel() + np.random.randn(100) * 0.5  # Sinusoidal relationship with noise

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Underfitting model - Linear Regression
model_underfit = LinearRegression()
model_underfit.fit(X_train, y_train)

# Overfitting model - High degree Polynomial Regression
model_overfit = make_pipeline(PolynomialFeatures(degree=15), LinearRegression())
model_overfit.fit(X_train, y_train)

# Correctly fitted model - Moderate degree Polynomial Regression
model_correct = make_pipeline(PolynomialFeatures(degree=3), LinearRegression())
model_correct.fit(X_train, y_train)
```
